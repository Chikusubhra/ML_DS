```python
# import dependecies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
column_name=["CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD","TAX","PTRATIO","B","LSTAT","PRICE"]
```

```python
house_data=pd.read_csv("housing.csv",header=None,delimiter=r"\s+",names=column_name)
house_data.head()
```

```python
house_data.shape
```

```python
house_data.isnull().sum()
```

```python
house_data.describe()
```

```python
house_data.isnull().sum()
```

```python
# co-relation
correlation=house_data.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt=".1f",annot=True,annot_kws={"size":10},cmap="Blues")
```

```python
# splitting data
X=house_data.drop(["PRICE"],axis=1)
Y=house_data["PRICE"]
```

```python
print(X)
```

```python
print(Y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,x_train.shape,x_test.shape)
```

```python
# model training
model=XGBRegressor()
```

```python
model.fit(x_train,y_train)
```

```python
# accuracy
train_data_predict=model.predict(x_train)
```

```python
print(train_data_predict)
```

```python
# error(rse,mae)
score_1=metrics.r2_score(y_train,train_data_predict)
score_2=metrics.mean_absolute_error(y_train,train_data_predict)
```

```python
print("R squared error: ",score_1)
print("mean absolute error: ",score_2)
```

```python
plt.scatter(y_train,train_data_predict)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price s Predicted Prices")
plt.show()
```

```python
# testing
test_data_predict=model.predict(x_test)
```

```python
# error(rse,mae)
score_1=metrics.r2_score(y_test,test_data_predict)
score_2=metrics.mean_absolute_error(y_test,test_data_predict)
```

```python
print("R squared error: ",score_1)
print("mean absolute error: ",score_2)
```
